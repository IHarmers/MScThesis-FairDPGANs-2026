## Plots of Results

```python
# Plotting privacy gains acquired from membership inference attacks performed on all our models.
labels = ["TabFairGAN", "DP-GAN", "DP-CTGAN", "FairDPGAN (Dem)", "FairDPGAN (Dis)", "FairDPGAN (Dem + Dis)", "FairDPGAN (Clf. Dem)", 
          "FairDPGAN (Clf. Dis)", "FairDPGAN (Clf. Eqopp)", "FairDPGAN (Clf. All)"]
colors = ["blue", "purple", "lime", "forestgreen", "darkcyan", "mediumaquamarine", "darkviolet", "violet", "palevioletred", "mediumpurple"]
tick_place = [0, 0.25, 0.68, 1.2]
epsi = [1, 2, 5]
markers = ["o", "s", "P", "v", "*"]

for r in range(0, 3):
    # Create a figure for the aggregated plot.
    fig, axes = plt.subplots(1, 3, figsize = (16, 6.0))  # 1 x 3 grid
    for e in range(len(epsi)):
        if r == 0:
            dataset = "Adult"
            TF = TF_adult
            DP = [DP1_adult, DP2_adult, DP5_adult]
            DPC = [DPC1_adult, DPC2_adult, DPC5_adult]
            dem = [dem1_adult, dem2_adult, dem5_adult]
            dis = [dis1_adult, dis2_adult, dis5_adult]
            demdi = [demdi1_adult, demdi2_adult, demdi5_adult]
            clf_dem = [clf_dem1_adult, clf_dem2_adult, clf_dem5_adult]
            clf_dis = [clf_dis1_adult, clf_dis2_adult, clf_dis5_adult]
            clf_eqopp = [clf_eqopp1_adult, clf_eqopp2_adult, clf_eqopp5_adult]
            clf_all = [clf_all1_adult, clf_all2_adult, clf_all5_adult]
            results = [TF, DP[e], DPC[e], dem[e], dis[e], demdi[e], clf_dem[e], clf_dis[e], clf_eqopp[e], clf_all[e]]
        elif r == 1:
            dataset = "Bank"
            TF = TF_bank
            DP = [DP1_bank, DP2_bank, DP5_bank]
            DPC = [DPC1_bank, DPC2_bank, DPC5_bank]
            dem = [dem1_bank, dem2_bank, dem5_bank]
            dis = [dis1_bank, dis2_bank, dis5_bank]
            demdi = [demdi1_bank, demdi2_bank, demdi5_bank]
            clf_dem = [clf_dem1_bank, clf_dem2_bank, clf_dem5_bank]
            clf_dis = [clf_dis1_bank, clf_dis2_bank, clf_dis5_bank]
            clf_eqopp = [clf_eqopp1_bank, clf_eqopp2_bank, clf_eqopp5_bank]
            clf_all = [clf_all1_bank, clf_all2_bank, clf_all5_bank]
            results = [TF, DP[e], DPC[e], dem[e], dis[e], demdi[e], clf_dem[e], clf_dis[e], clf_eqopp[e], clf_all[e]]
        else:
            dataset = "Credit"
            TF = TF_credit
            DP = [DP1_credit, DP2_credit, DP5_credit]
            DPC = [DPC1_credit, DPC2_credit, DPC5_credit]
            dem = [dem1_credit, dem2_credit, dem5_credit]
            dis = [dis1_credit, dis2_credit, dis5_credit]
            demdi = [demdi1_credit, demdi2_credit, demdi5_credit]
            clf_dem = [clf_dem1_credit, clf_dem2_credit, clf_dem5_credit]
            clf_dis = [clf_dis1_credit, clf_dis2_credit, clf_dis5_credit]
            clf_eqopp = [clf_eqopp1_credit, clf_eqopp2_credit, clf_eqopp5_credit]
            clf_all = [clf_all1_credit, clf_all2_credit, clf_all5_credit]
            results = [TF, DP[e], DPC[e], dem[e], dis[e], demdi[e], clf_dem[e], clf_dis[e], clf_eqopp[e], clf_all[e]]

        # Customizing the plot.
        ax = axes[e]
        ax.set_title(f"{dataset} ($\epsilon = {epsi[e]}$)")
        if e == 0:
            ax.set_ylabel("Privacy Gain")
        ax.set_aspect('auto')
        ax.set_xlim([-0.25, 1.5])
        ax.set_ylim([-0.05, 2.05])
        ax.xaxis.set_major_locator(ticker.FixedLocator(tick_place))
        ax.xaxis.set_major_formatter(ticker.FixedFormatter(["TF", "DPs", "FDPs (D)", "FDPs (C)"]))
        ax.grid(True, linestyle = '--', alpha = 0.6)
    
        # Plotting the zero line.
        ax.axhline(0, 0, 5, ls = "-", c = "red", zorder = 0)
            
        for p in range(len(tick_place)):
            if p == 0:
                for m in range(len(markers)):
                    ax.scatter([tick_place[p]], results[p].iloc[m], color = colors[p], marker = markers[m], s = 62,
                               edgecolors = 'black')   # TabFairGAN
            elif p == 1:
                for m in range(len(markers)):
                    ax.scatter([tick_place[p] - 0.05], results[p].iloc[m], color = colors[p], marker = markers[m], s = 62,
                               edgecolors = 'black')   # DP-GAN
                    ax.scatter([tick_place[p] + 0.05], results[p + 1].iloc[m], color = colors[p + 1], marker = markers[m], s = 62,
                               edgecolors = 'black')   # DP-CTGAN
            elif p == 2:
                for m in range(len(markers)):
                    ax.scatter([tick_place[p] - 0.07], results[p + 1].iloc[m], color = colors[p + 1], marker = markers[m], s = 62,
                               edgecolors = 'black')   # Dem
                    ax.scatter([tick_place[p]], results[p + 2].iloc[m], color = colors[p + 2], marker = markers[m], s = 62,
                               edgecolors = 'black')   # Dis
                    ax.scatter([tick_place[p] + 0.07], results[p + 3].iloc[m], color = colors[p + 3], marker = markers[m], s = 62,
                               edgecolors = 'black')   # Dem + Dis
            else: 
                if e != 3:
                    for m in range(len(markers)):
                        ax.scatter([tick_place[p] - 0.13], results[p + 3].iloc[m], color = colors[p + 3], marker = markers[m], s = 62,
                                   edgecolors = 'black')   # Clf. Dem
                        ax.scatter([tick_place[p] - 0.05], results[p + 4].iloc[m], color = colors[p + 4], marker = markers[m], s = 62,
                                   edgecolors = 'black')   # Clf. Dis
                        ax.scatter([tick_place[p] + 0.05], results[p + 5].iloc[m], color = colors[p + 5], marker = markers[m], s = 62,
                                   edgecolors = 'black')   # Clf. Eqopp
                        ax.scatter([tick_place[p] + 0.13], results[p + 6].iloc[m], color = colors[p + 6], marker = markers[m], s = 62,
                                   edgecolors = 'black')   # Clf. All
                    else:
                        pass

        if r == 0:
            plt.figlegend(handles = [mpatches.Patch(color = colors[i], label = labels[i]) for i in range(len(labels))], ncol = 4, 
                          loc = "upper center", bbox_to_anchor = (0.5, 1.15))
        
    plt.tight_layout(rect = [0, 0, 1, 0.95])
plt.show()
```
