```python
# Importing libraries.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import matplotlib
params = {'axes.titlesize':'18',
          'xtick.labelsize':'15',
          'ytick.labelsize':'15',
          'font.size':'16',
          'legend.fontsize':'medium',
          'lines.linewidth':'2.0',
          'font.weight':'normal',
          'lines.markersize':'5',
          'text.latex.preamble': r'\usepackage{amsfonts}',
          'hatch.linewidth': '0.3',
          'hatch.color': "grey"
          }
matplotlib.rcParams.update(params)
plt.rcParams["mathtext.fontset"] = "cm"
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

def tf(data, counts):
    """This function transforms the given data to a percentile-representation when provided with a total 'count'."""
    return np.round(((np.array(data)/counts)*100), 1)
```

## Summary Plots of Utility Metrics

```python
# Summary plots of utility (AUROC and accuracy) improvements: Fair DP-GANs versus DP-GAN. 
counts_util = 24

fdis_util = tf([5, 3], counts_util)
f075_util = tf([5, 4], counts_util)
f100_util = tf([6, 4], counts_util)
f125_util = tf([6, 5], counts_util)
f150_util = tf([5, 4], counts_util)
results_util = [fdis_util, f075_util, f100_util, f125_util, f150_util]

fdis_fdata = tf([2, 2, 4], counts_util)
f075_fdata = tf([4, 2, 3], counts_util)
f100_fdata = tf([3, 4, 3], counts_util)
f125_fdata = tf([3, 2, 6], counts_util)
f150_fdata = tf([2, 4, 3], counts_util)
results_fdata = [fdis_fdata, f075_fdata, f100_fdata, f125_fdata, f150_fdata]

fig, ax = plt.subplots(1, 2, figsize = (16, 5.0))
ax[0].set_ylim(-0.4, 4.4)
ax[1].set_ylim(-0.4, 4.4)
ax[0].axvline(x = 50, color = "black", zorder = 0)
ax[1].axvline(x = 50, color = "black", zorder = 0)

rect1a = mpatches.Rectangle(xy = (-0.5, -1), width = 50, height = 10, color = 'red', alpha = 0.1, ec = 'red')
ax[0].add_patch(rect1a)
rect1b = mpatches.Rectangle(xy = (-0.5, -1), width = 50, height = 10, color = 'red', alpha = 0.1, ec = 'red')
ax[1].add_patch(rect1b)
rect2a = mpatches.Rectangle(xy = (50, -1), width = 50, height = 10, color = 'green', alpha = 0.1, ec = 'green')
ax[0].add_patch(rect2a)
rect2b = mpatches.Rectangle(xy = (50, -1), width = 50, height = 10, color = 'green', alpha = 0.1, ec = 'green')
ax[1].add_patch(rect2b)

positions = [0, 1, 2, 3, 4]
width = 0.4
colors = ["cyan", "gold", "thistle", "darkkhaki", "powderblue"]
labels = ["Accuracy", "AUROC", "Adult", "Bank", "Credit"]
hatches = [None, None, "/" , "--" , "x"]
legd_hatch = [None, None, "//" , "---" , "xx"]

for a in range(2):
    if a == 0:
        results = results_util
    else:
        results = results_fdata

    for r in range(len(results)):
        s = 0
        for c in range(len(results[r])):
            hbars = ax[a].barh(positions[r], results[r][c], width, align = "center", left = [s], color = [colors[(c + a * 2)]]*2, 
                               hatch = [hatches[(c + a * 2)]]*2, 
                              )
            ax[a].bar_label(hbars, label_type = "center", fontsize = 17, color = "black")
            s += results[r][c]
            
    ax[a].set_xlabel("\%")
    ax[a].set_xlim([0, 100])
    ax[a].grid(True, axis = "x", linestyle = '--', alpha = 0.6)

ax[0].set_yticks(positions, labels = ["$Dis$ ($\lambda_f = 0.5$)", "$Dis$ ($\lambda_f = 0.75$)", "$Dis$ ($\lambda_f = 1.0$)", 
                                      "$Dis$ ($\lambda_f = 1.25$)", "$Dis$ ($\lambda_f = 1.5$)"])
ax[1].set_yticks(positions, labels = [" "]*5)
ax[0].invert_yaxis()  # labels read top-to-bottom
ax[1].invert_yaxis()

fig.suptitle(r"Classifier Utility Improvements Across $\epsilon \in \{1, 2, 5, 8\}$: FairDPGANs Versus DP-GAN ($\bf{T}$)")
plt.figlegend(handles = [mpatches.Patch(facecolor = colors[i], label = labels[i], hatch = legd_hatch[i]) for i in range(len(colors))], ncol = 6, 
              loc = "upper center", bbox_to_anchor = (0.47, 1.13))

plt.figtext(0.40, 0.80, "vs. DP-GAN", size = 14)
plt.figtext(0.40, 0.645, "vs. DP-GAN", size = 14)
plt.figtext(0.40, 0.49, "vs. DP-GAN", size = 14)
plt.figtext(0.40, 0.33, "vs. DP-GAN", size = 14)
plt.figtext(0.40, 0.17, "vs. DP-GAN", size = 14)

plt.figtext(0.823, 0.80, "vs. DP-GAN", size = 14)
plt.figtext(0.823, 0.645, "vs. DP-GAN", size = 14)
plt.figtext(0.823, 0.49, "vs. DP-GAN", size = 14)
plt.figtext(0.823, 0.33, "vs. DP-GAN", size = 14)
plt.figtext(0.823, 0.17, "vs. DP-GAN", size = 14)

plt.show()
```
