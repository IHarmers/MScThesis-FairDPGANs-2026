```python
# Importing libraries.
import numpy as np
import pandas as pd
from ydata_synthetic.synthesizers import ModelParameters, TrainParameters
from ydata_synthetic.synthesizers.regular import RegularSynthesizer
```

```python
# Importing train set.
adult_train = pd.read_csv("./train-test-datasets/adult/adult_train.csv")

num_cols = adult_train.select_dtypes(exclude=['object']).columns.to_list()
cat_cols = adult_train.select_dtypes(include=['object']).columns.to_list()
print(num_cols, cat_cols)
adult_train.describe()
```

```python
# Defining the training parameters.
# For more information see https://docs.synthetic.ydata.ai/1.3/synthetic_data/single_table/wgangp_example/ 
# (accessed June 28, 2025).

noise_dim = 128
dim = 128
batch_size = 500

log_step = 100
epochs = 500 + 1
learning_rate = [5e-4, 3e-3]
beta_1 = 0.5
beta_2 = 0.9

gan_args = ModelParameters(batch_size = batch_size,
                           lr = learning_rate,
                           betas = (beta_1, beta_2),
                           noise_dim = noise_dim,
                           layers_dim = dim)

train_args = TrainParameters(epochs = epochs,
                             sample_interval = log_step)
```

```python
# Synthesizing the dataset with WGAN-GP.
synth = RegularSynthesizer(modelname = 'wgangp', model_parameters = gan_args, n_critic = 2)
synth.fit(adult_train, train_args, num_cols, cat_cols, plot_losses = True)
```

```python
# Generating a synthetic dataset with the same amount of rows as the original training dataset.
sample = synth.sample(adult_train.shape[0])
```
