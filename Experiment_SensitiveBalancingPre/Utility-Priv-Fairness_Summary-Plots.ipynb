```python
# Importing libraries.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import matplotlib
params = {'axes.titlesize':'18',
          'xtick.labelsize':'15',
          'ytick.labelsize':'15',
          'font.size':'16',
          'legend.fontsize':'medium',
          'lines.linewidth':'2.0',
          'font.weight':'normal',
          'lines.markersize':'5',
          'text.latex.preamble': r'\usepackage{amsfonts}',
          'hatch.linewidth': '0.3',
          'hatch.color': "grey"
          }
matplotlib.rcParams.update(params)
plt.rcParams["mathtext.fontset"] = "cm"
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

def tf(data, counts):
    """This function transforms the given data to a percentile-representation when provided with a total 'count'."""
    return np.round(((np.array(data)/counts)*100), 1)
```

## Summary Plots of Utility Metrics

```python
# Summary plots of utility (AUROC and accuracy) improvements: Fair DP-GANs versus DP-CTGAN & DP-GAN. 
counts_util = 24

fdem_util = tf([6, 6], counts_util)
fdis_util = tf([4, 4], counts_util)
fdemdi_util = tf([4, 4], counts_util)
results_util = [fdem_util, fdis_util, fdemdi_util]

fdem_util_DPC = tf([9, 9], counts_util)
fdis_util_DPC = tf([8, 8], counts_util)
fdemdi_util_DPC = tf([9, 9], counts_util)
results_util_DPC = [fdem_util_DPC, fdis_util_DPC, fdemdi_util_DPC]

fdem_fdata = tf([2, 4, 6], counts_util)
fdis_fdata = tf([4, 2, 2], counts_util)
fdemdi_fdata = tf([2, 4, 2], counts_util)
results_fdata = [fdem_fdata, fdis_fdata, fdemdi_fdata]

fdem_fdata_DPC = tf([6, 6, 6], counts_util)
fdis_fdata_DPC = tf([6, 6, 4], counts_util)
fdemdi_fdata_DPC = tf([8, 6, 4], counts_util)
results_fdata_DPC = [fdem_fdata_DPC, fdis_fdata_DPC, fdemdi_fdata_DPC]

fig, ax = plt.subplots(1, 2, figsize = (16, 5.0))
ax[0].set_ylim(-0.56, 2.55)
ax[1].set_ylim(-0.56, 2.55)
ax[0].axvline(x = 50, color = "black", zorder = 0)
ax[1].axvline(x = 50, color = "black", zorder = 0)

rect1a = mpatches.Rectangle(xy = (-0.5, -1), width = 50, height = 10, color = 'red', alpha = 0.1, ec = 'red')
ax[0].add_patch(rect1a)
rect1b = mpatches.Rectangle(xy = (-0.5, -1), width = 50, height = 10, color = 'red', alpha = 0.1, ec = 'red')
ax[1].add_patch(rect1b)
rect2a = mpatches.Rectangle(xy = (50, -1), width = 50, height = 10, color = 'green', alpha = 0.1, ec = 'green')
ax[0].add_patch(rect2a)
rect2b = mpatches.Rectangle(xy = (50, -1), width = 50, height = 10, color = 'green', alpha = 0.1, ec = 'green')
ax[1].add_patch(rect2b)

positions = [0, 1, 2]
width = 0.4
colors = ["cyan", "gold", "thistle", "darkkhaki", "powderblue"]
labels = ["Accuracy", "AUROC", "Adult", "Bank", "Credit"]
hatches = [None, None, "/" , "--" , "x"]
legd_hatch = [None, None, "//" , "---" , "xx"]

for a in range(2):
    if a == 0:
        results = results_util
        results_DPC = results_util_DPC
    else:
        results = results_fdata
        results_DPC = results_fdata_DPC

    for r in range(len(results)):
        s1 = 0
        s2 = 0
        for c in range(len(results[r])):
            hbars = ax[a].barh([positions[r] - 0.55*width, positions[r] + 0.55*width], [results_DPC[r][c], results[r][c]], width, align = "center", 
                               left = [s1, s2], color = [colors[(c + a * 2)]]*2, hatch = [hatches[(c + a * 2)]]*2, 
                              )
            ax[a].bar_label(hbars, label_type = "center", fontsize = 20, color = "black")
            s1 += results_DPC[r][c]
            s2 += results[r][c]
            
    ax[a].set_xlabel("\%")
    ax[a].set_xlim([0, 100])
    ax[a].grid(True, axis = "x", linestyle = '--', alpha = 0.6)

ax[0].set_yticks(positions, labels = ["$Dem$", "$Dis$", "$Dem$ + $Dis$"])
ax[1].set_yticks(positions, labels = [" "]*3)
ax[0].invert_yaxis()  # labels read top-to-bottom
ax[1].invert_yaxis()

fig.suptitle(r"Classifier Utility Improvements Across $\epsilon \in \{1, 2, 5, 8\}$: FairDPGANs Versus DP-CTGAN \& DP-GAN ($\bf{S}$)")
plt.figlegend(handles = [mpatches.Patch(facecolor = colors[i], label = labels[i], hatch = legd_hatch[i]) for i in range(len(colors))], ncol = 6, 
              loc = "upper center", bbox_to_anchor = (0.47, 1.13))

plt.figtext(0.387, 0.78,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.404, 0.67,  "vs. DP-GAN", size = 14)
plt.figtext(0.387, 0.53,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.404, 0.425,  "vs. DP-GAN", size = 14)
plt.figtext(0.387, 0.28,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.404, 0.177,  "vs. DP-GAN", size = 14)

plt.figtext(0.809, 0.78,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.826, 0.67,  "vs. DP-GAN", size = 14)
plt.figtext(0.809, 0.53,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.826, 0.425,  "vs. DP-GAN", size = 14)
plt.figtext(0.809, 0.28,  "vs. DP-CTGAN", size = 14)
plt.figtext(0.826, 0.177,  "vs. DP-GAN", size = 14)

plt.show()
```
